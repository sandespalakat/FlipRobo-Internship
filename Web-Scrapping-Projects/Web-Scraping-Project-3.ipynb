```python
from selenium import webdriver
import time
import pandas as pd
import numpy as np
import urllib
import os
```

```python
#initialize web driver
driver = webdriver.Chrome(executable_path = r"C:\Users\DEVASREE\Downloads\chromedriver.exe")
url = "https://www.amazon.in/"
driver.get(url)
time.sleep(3)
```

```python
#accepts input from user and search on amazon.in
user_uqery = input("Enter the product you want to search: ")
search_field = driver.find_element_by_id("twotabsearchtextbox")
search_field.clear()
search_field.send_keys(user_uqery)
search_button = driver.find_element_by_id("nav-search-submit-button")
search_button.click()
```

```python
time.sleep(5)
#initializing lists
brands = []
titles = []
ratings = []
no_ratings = []
prices = []
returns = []
exp_delis = []
avails = []
others = []
urls = []

#get product urls for first 3 pages
for i in range(0,3):
    time.sleep(5)
    #get the individual product div
    divs = driver.find_elements_by_xpath("//div[@data-component-type='s-search-result']")
    for div in divs:
        try:
            #appending product url to the list
            urls.append(div.find_element_by_class_name("s-no-outline").get_attribute("href"))
        except Exception as e:
            urls.append("--")
    try:
        #get the url of the next page and redirects to next page
        next_page = driver.find_elements_by_xpath("//li[@class='a-last']")[-1]
        driver.get(next_page.find_element_by_tag_name('a').get_attribute("href"))
        time.sleep(5)
    except Exception as e:
        #if products less than 3 pages, breaks the loop
        break
```

```python
#looping through the urls to get details of the product
for url in urls:
    if url != "--":
        driver.get(url)
        time.sleep(10)
        try:
            brand = driver.find_element_by_id("bylineInfo").text.strip()
            if "Visit" in brand:
                for i in ["Visit", "the", "Store"]:
                    brand = brand.replace(i, "")
            else:
                brand = brand.replace("Brand:", "")
            brands.append(brand)
        except:
            brands.append("--")
        try:
            name = driver.find_element_by_id("productTitle").text
            titles.append(name)
        except:
            titles.append("--")
        try:
            rating = driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']").text
            ratings.append(rating)
        except:
            ratings.append("--")
        try:
            number = driver.find_element_by_id("acrCustomerReviewText").text.split(" ")[0]
            no_ratings.append(number)
        except:
            no_ratings.append("--")
        try:
            price = driver.find_element_by_id("priceblock_ourprice").text.split("₹")[1]
            prices.append(price)
        except:
            prices.append("--")
        try:
            retrn = driver.find_element_by_xpath("//div[@class='a-row icon-farm-wrapper']").text
            if "replacement" in retrn.lower():
                returns.append("Yes")
            else:
                returns.append("No")
        except Exception as e:
            returns.append("--")
        try:
            exp = driver.find_element_by_id("ddmDeliveryMessage").text
            exp_delis.append(exp)
        except:
            exp_delis.append("--")
        if "currently unavailable" in str(driver.page_source).lower():
            availability = 'No'
        else:
            availability = 'Yes'
        try:
            availability = driver.find_element_by_id("availability").text
            if "stock" in availability.lower():
                avails.append("Yes")
            else:
                avails.append("No")
        except:
            avails.append("--")
        try:
            other = driver.find_element_by_id("feature-bullets").text
            others.append(other)
        except:
            avails.append("--")
products = pd.DataFrame({"Brand":brands,"Title":titles,"Rating":ratings,
                "No of Ratings":no_ratings, "Price":prices,
                        "Return":returns, "Delivery":exp_delis,
                        "Availability":avails,"Others":others,"URL":urls})
print(products)
        
```

```python
#saving the details as csv
products.to_csv("Guitars.csv", sep='\t')
```

```python
for keyword in ["fruits", "cars", "Machine Learning"]:
    search = driver.find_element_by_name("q")
    search.clear()
    search.send_keys(keyword)
    driver.find_element_by_xpath("//button[@aria-label='Google Search']").click()
    #scrolling down to load more images
    time.sleep(5)
    value = 0
    for i in range(0,4):
        driver.execute_script("scrollBy(" + str(value) + ", +1000);")
        value = value + 2000
        time.sleep(2)
    time.sleep(6)
    images = driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
    j = 1
    for image in images:
        #if any url is not downloadable, loop until downloads exactly 100 images
        if j <= 100:
            src = image.get_attribute("src")
            folder = keyword
            if not os.path.exists(folder):
                os.makedirs(folder)
            try:
                if src != None:
                    urllib.request.urlretrieve(str(src), os.path.join( folder +"/"+ keyword + str(j) + ".jpg"))
                    j = j + 1
            except Exception as e:
                print(str(e))
            
        
```

```python
#initialize web driver
driver = webdriver.Firefox(executable_path = r"C:/Users/DEVASREE/Downloads/geckodriver.exe")
url = "https://trak.in/india-startup-funding-investment-2015/"
driver.get(url)
time.sleep(3)
```

```python
#get all the heading 
headings = driver.find_elements_by_class_name("tablepress-table-name")
#get all tables
tables = driver.find_elements_by_class_name("row-hover")
#get all select button(for viewing max no of rows)
options = driver.find_elements_by_tag_name("select")
#initializing the info lists
dates = []
start_ups = []
industires = []
sub_verticals = []
cities = []
investors = []
invest_types = []
amounts = []
#set scrap value =false initially to scrap only between july 2020 and sept 2020
scrap = False
for head,table,option in zip(headings, tables, options):
    if "September, 2020" in head.text:
        scrap = True
    if "June, 2020" in head.text:
        break
    if scrap:
        #select the no of rows to 100
        option.find_elements_by_tag_name("option")[-1].click()
        time.sleep(10)
        #finding the rows of the particular table
        trs = table.find_elements_by_tag_name("tr")
        for tr in trs:
            #finding and appending the data from the curresponding row
            dates.append(tr.find_element_by_class_name("column-2").text)
            start_ups.append(tr.find_element_by_class_name("column-3").text)
            industires.append(tr.find_element_by_class_name("column-4").text)
            sub_verticals.append(tr.find_element_by_class_name("column-5").text)
            cities.append(tr.find_element_by_class_name("column-6").text)
            investors.append(tr.find_element_by_class_name("column-7").text)
            invest_types.append(tr.find_element_by_class_name("column-8").text)
            amounts.append(tr.find_element_by_class_name("column-9").text)
            
# creating datafrrame
fundings = pd.DataFrame({"Date:":dates, "Startup Name":start_ups, "Industry": industires, 
                         "Sub Vertical": sub_verticals, "City":cities, "Investor": investors,
                        "Invest Type": invest_types, "Amount": amounts})
fundings
                
```

```python
#initialize web driver
driver = webdriver.Firefox(executable_path = r"C:/Users/DEVASREE/Downloads/geckodriver.exe")
base_url = "https://www.digit.in"
driver.get(url)
time.sleep(3)
```

```python
#go to search url
driver.get(base_url+"/search/?keyword=Gaming%20laptops")
divs = driver.find_elements_by_class_name("searchPage")
urls = []
for div in divs:
    pdct = div.find_element_by_tag_name("p").text
    #append to url iff the item is product not an article
    if "PRODUCT" in pdct:
        href = div.find_element_by_tag_name("a").get_attribute("href")
        urls.append(href)      
driver.get(base_url+"/search?keyword=Gaming%20laptops&page=2")
time.sleep(5)
divs = driver.find_elements_by_class_name("searchPage")
for div in divs:
    pdct = div.find_element_by_tag_name("p").text
    if "PRODUCT" in pdct:
        href = div.find_element_by_tag_name("a").get_attribute("href")
        urls.append(href)
#initializing the info lists
names = []
launched = []
released = []
os = []
displays = []
processors = []
prices = []
#looping through the urls to get full information
j = 1
for url in urls:
    driver.get(url)
    time.sleep(8)
    bread_crumb = driver.find_element_by_class_name("breadcrumb").text
    #check the product is a laptop
    if "laptop" in bread_crumb.lower():
        #check for 10 products
        if j <= 10:
            try:
                name = driver.find_element_by_class_name("heading-wraper").find_element_by_tag_name("h1").text
                names.append(name)
            except:
                names.append("--")
            try:
                launch = driver.find_element_by_class_name("status-bar").find_element_by_class_name("Block-status").find_element_by_tag_name("b").get_attribute("innerHTML")
                launched.append(launch)
            except Exception as e:
                print(str(e))
                launched.append("--")
            try:
                release = driver.find_element_by_class_name("status-bar").find_elements_by_class_name("Block-status")[1].find_element_by_tag_name("b").get_attribute("innerHTML")
                released.append(release)
            except Exception as e:
                print(str(e))
                released.append("--")
            try:
                system = driver.find_elements_by_class_name("value")[0].text
                os.append(system)
            except Exception as e:
                print(str(e))
                os.append("--")
            try:
                display = driver.find_elements_by_class_name("value")[1].text
                displays.append(display)
            except Exception as e:
                print(str(e))
                displays.append("--")
            try:
                processor = driver.find_elements_by_class_name("value")[2].text
                processors.append(processor)
            except Exception as e:
                print(str(e))
                processors.append("--")
            try:
                price = driver.find_element_by_class_name("Block-price").text.replace("Price :", "")
                prices.append(price)
            except Exception as e:
                print(str(e))
                prices.append("--")
            j = j + 1
        else:
            break
#creating dataframe
laptops = pd.DataFrame({"Name":names, "Status": launched, "Release": released, "OS":os, "Display":displays
                       ,"Processor":processors, "Price":prices})
laptops
    
```
